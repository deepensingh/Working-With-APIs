# Task - 1

```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=b94529721acdbbd67f18655c174254c6')
```

```python
response
```

```python
temp_df = pd.DataFrame(response.json()['results']).head(5)[['id','title','overview','popularity','vote_average','vote_count']]
```

```python
temp_df.head(10)
```

```python
df = pd.DataFrame()
```

```python
for i in range(1,430):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=b94529721acdbbd67f18655c174254c6&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results']).head(5)[['id','title','overview','popularity','vote_average','vote_count']]
    df = df.append(temp_df, ignore_index=True )
    
    
    
    
    
```
